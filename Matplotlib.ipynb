## Visualizing using Pandas (in-built plot function)

### Creating Sample Data

```python
# Data contains quantity and price for two stores A & B for 30 days

q=np.random.randint(1,20,60).reshape(30,2)
p=np.random.randint(500,1000,60).reshape(30,2)
data=np.hstack((q,p))
df=pd.DataFrame(data,columns=['Quantity_A','Quantity_B','Price_A','Price_B'])
df['Revenue_A']=df['Quantity_A']*df['Price_A']
df['Revenue_B']=df['Quantity_B']*df['Price_B']
df.index.names=['Days']
df
```

```python
# plotting dataframe

# dataframe has an in-built function plot()
df.plot(figsize=(11,6)); # parameter figsize takes a tuple (width, height) as an input for the figure's size
```

```python
# using parameter subplots=True, each graph is plotted independently
df.plot(subplots=True,figsize=(11,10));

# Note that semicolons is added in the end in order to avoid the printing of object name
plt.tight_layout() # tight_layout() provides proper spacing between multiple graphs
```

```python
# layout parameter (rows,columns) re-arranges the subplots
df.plot(subplots=True,layout=(3,2),figsize=(13,10));  
plt.tight_layout()

# For each of these subplots, the x-axis (days) is shared 
```

```python
# in order to plot specific columns, example: Revenue_A & Revenue_B
df[['Revenue_A','Revenue_B']].plot(figsize=(10,4));
plt.tight_layout()
```

## Visualizing using Matplotlib

### Defining Figure

```python
x = np.arange(1,11) 
y = (2 * x + 5)*np.cos(x) 
z=np.sin(x)

plt.figure(figsize=(11,6)) # figure provides the area where the graph and other elements can be plotted 
plt.title("$(2x+5)cos(x)$ and $sin(x)$") #  $ symbols are used for enclosing mathematical expressions as Latex
plt.xlabel("x axis") 
plt.ylabel("y axis")

# values are plotted using plot()
plt.plot(x,y,label="$(2x+5)cos(x)$") 
plt.plot(x,z,'ro-',label="$sin(x)$") # 'o-' denotes how the graph is printed and r represents red color
plt.legend(); # legend() is used to display labels
```

```python
x = np.linspace(0, 2*np.pi, 100) 
y = np.sin(x)
z = np.cos(x) 

plt.figure(figsize=(11,6))
plt.title("sin(x) and cos(x)") 
plt.xlabel("x axis") 
plt.ylabel("y axis") 

# formatting the axis
plt.tick_params(axis='y',color='red',labelcolor='green',labelsize=10)
plt.tick_params(axis='x',labelcolor='red',labelsize=10)

# plotting both the functions together in a same graph
plt.plot(x,y,label="sin(x)")
plt.plot(x,z,'--m',label="cos(x)") # '--' denotes how the graph is printed and m represents magenta color
plt.legend(framealpha=0.5) # parameter framealpha controls the visibility of the legend box
plt.show() # graphical representation is displayed (works similar to the print())

# Save the figure (uncomment to save)
# plt.savefig('sin_cos.png')
```

```python
# plotting both graphs individually using subplot()

plt.figure(figsize=(15,6))
plt.subplot(1,2,1) # subplot(row, column, cell)
plt.plot(x,y,label="sin(x)")
plt.title("sin(x)")
plt.axhline(0,linestyle='--',color='green') # plotting horizontal line at y=0
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,z,'--m',label="cos(x)")
plt.axhline(0,linestyle='--',color='green')
plt.title("cos(x)") 
plt.legend();
```

```python
# customizing layout

s = np.linspace(-2*np.pi, 2*np.pi, 100) 
a = np.sin(s)
b = np.cos(s) 
c = np.tan(s)
d = np.exp2(s)

ax1=plt.subplot2grid((2,3),(0,0),colspan=2)
ax1.plot(s,d)
ax1.set_title('exp2(x)')

ax2=plt.subplot2grid((2,3),(0,2),rowspan=2)
ax2.plot(s,c)
ax2.set_title('tan(x)')

ax3=plt.subplot2grid((2,3),(1,0))
ax3.plot(s,a)
ax3.set_title('sin(x)')

ax4=plt.subplot2grid((2,3),(1,1))
ax4.plot(s,b)
ax4.set_title('cos(x)')


plt.tight_layout()
```

```python
# creating a canvas inside another canvas

# canvas 1
fig1=plt.figure(figsize=(11,5)) 
axes1=fig1.add_axes([0,0,1,1]) 
axes1.set_xlabel("X1 Axis")
axes1.set_ylabel("Y1 Axis")
axes1.set_title("Outer Plot")

# canvas 2
axes2=fig1.add_axes([0.2,0.2,0.6,0.6])
axes2.set_xlabel("X2 Axis")
axes2.set_ylabel("Y2 Axis")
axes2.set_title("Inner Plot")

#plotting
axes1.plot(np.sin((np.linspace(0,2*np.pi,100))),label="$sin(x)$",color='orange')
axes1.legend()
axes2.plot(np.cos((np.linspace(0,2*np.pi,100))),label="$cos(x)$",linestyle='--')
axes2.legend();
```

### Plotting Shapes

```python
import matplotlib.patches as patches

fig=plt.figure(figsize=(11,10))



# Figure_1
ax1=fig.add_subplot(321)
ax1.set_aspect(aspect='equal') # setting aspect of the figure for proper visualization, else x-axis appears longer

# add_patch() allows to introduce shapes on the graph
ax1.add_patch(patches.Rectangle((0.1,0.1),0.8,0.2)) # Rectangle((x,y),width, height)
ax1.add_patch(patches.Rectangle((0.1,0.5),0.3,0.3,fill=False,hatch='x',edgecolor='red',linestyle='-.')) # square

# hatch represents inside pattern, edgecolor represents color of edge, facecolor represents fill color
ax1.add_patch(patches.Rectangle((0.6,0.5),0.3,0.3,facecolor='brown',hatch='.',edgecolor='yellow'))



# Figure_2
ax2=fig.add_subplot(322)
ax2.set_aspect(aspect='equal')

for p in [
    patches.Rectangle((0.1,0.1),0.2,0.6,fill=False,hatch='+',linestyle='--',edgecolor='purple'),
    patches.Rectangle((0.4,0.1),0.2,0.8,facecolor='orange',hatch='/',linestyle=':',edgecolor='blue'),
    patches.Rectangle((0.7,0.1),0.2,0.7,facecolor='green',hatch='o',alpha=0.6)
]:
    ax2.add_patch(p)
    


# Figure_3
ax3=fig.add_subplot(323)
ax3.set_aspect(aspect='equal')

for p in [
    patches.Arrow(0.5,0.5,0,0.3,width=0.2), # Arrow(x,y,dx,dy,width)
    patches.Arrow(0.5,0.5,0,-0.3,width=0.2,alpha=0.7),
    patches.Arrow(0.5,0.5,0.3,0,width=0.2),
    patches.Arrow(0.5,0.5,-0.3,0,width=0.2,alpha=0.7)
]:
    ax3.add_patch(p)
ax3.text(0.48,0.83,'N',fontsize=15)
ax3.text(0.48,0.13,'S',fontsize=15)
ax3.text(0.81,0.48,'E',fontsize=15)
ax3.text(0.12,0.48,'W',fontsize=15)


# Figure_4
ax4=fig.add_subplot(324)
ax4.set_aspect(aspect='equal')
for p in [
    patches.Circle((0.5,0.5),0.1,facecolor='red',hatch='*',linestyle='--',edgecolor='purple',alpha=0.6), # Circle((x,y),radius)
    patches.Circle((0.3,0.5),0.1,facecolor='pink',hatch='.',linestyle=':',edgecolor='blue'),
    patches.Circle((0.7,0.5),0.1,facecolor='orange',hatch='\\\\',linestyle='-.'),
    patches.Circle((0.5,0.7),0.1,hatch='/',alpha=0.7),
    patches.Circle((0.5,0.3),0.1,fill=False,hatch='||',edgecolor='green')
]:
    ax4.add_patch(p)



# Figure_5
ax5=fig.add_subplot(325)
polygon=patches.Polygon([[0.1,0.1],[0.3,0.9],[0.6,0.8],[0.7,0.2],[0.4,0.4]],fill=False) #by default, the polygon is closed
ax5.add_patch(polygon)



# Figure_6
ax6=fig.add_subplot(326)
polygon=patches.Polygon([[0.1,0.1],[0.3,0.9],[0.6,0.8],[0.7,0.2],[0.4,0.4]],fill=False,closed=False)
ax6.add_patch(polygon)

plt.tight_layout()
```

```python
from matplotlib.path import Path

fig,ax=plt.subplots(1,2,figsize=(11,4))
p1=patches.PathPatch(Path([(0.1,0.3),(0.2,0.1),(0.4,0.9),(0.5,0.4),(0.6,0.8),(0.8,0.1),(0.9,0.3)],
                      [Path.MOVETO,Path.LINETO,Path.LINETO,Path.LINETO,Path.LINETO,Path.LINETO,Path.LINETO]),fill=None)
# Path.MOVETO takes us to the point (0.1,0.3); Path.LINETO makes a line from the current point to the next point (0.2,0.1)
ax[0].add_patch(p1)

p2=patches.PathPatch(Path([(0.1,0.3),(0.2,0.1),(0.4,0.9),(0.5,0.4),(0.6,0.8),(0.8,0.1),(0.9,0.3)],
                      [Path.MOVETO,Path.LINETO,Path.LINETO,Path.LINETO,Path.LINETO,Path.LINETO,Path.CLOSEPOLY]),fill=None)
# Path.CLOSEPOLY closes the path to the first point (0.1,0.3), it ignores the last point (0.9,0.3)
ax[1].add_patch(p2);
```

```python
# Beizer Curve

from matplotlib.path import Path

fig,ax=plt.subplots(1,2,figsize=(11,5))

p=patches.PathPatch(Path([(0.1,0.3),(0.2,0.1),(0.4,0.9),(0.5,0.4),(0.6,0.8),(0.8,0.1),(0.9,0.3)],
                      [Path.MOVETO,Path.LINETO,Path.CURVE3,Path.LINETO,Path.CURVE3,Path.LINETO,Path.LINETO]),fill=None)
# Path.CURVE3 uses three points to plot a quadratic beizer curve
# (0.2,0.1),(0.5,0.4) are the endpoints for the first curve, (0.4,0.9) acts like a control point for this curve

p1=patches.PathPatch(Path([(0.1,0.3),(0.2,0.1),(0.4,0.9),(0.5,0.4),(0.6,0.8),(0.8,0.1),(0.9,0.3)],
                      [Path.MOVETO,Path.LINETO,Path.CURVE3,Path.LINETO,Path.CURVE3,Path.LINETO,Path.LINETO]),fill=None)

ax[0].add_patch(p)
ax[0].set_title('Beizer Curves')
ax[1].add_patch(p1)
ax[1].set_title('Beizer Curves illutrated using Annotations')

# annotations are used to represent the end & control points for the beizer curves
ax[1].annotate('Control Points',xy=(0.4,0.9),xytext=(0.1,0.65),arrowprops=dict(facecolor='g',alpha=0.3,shrink=0.1))
ax[1].annotate('Control Points',xy=(0.6,0.8),xytext=(0.1,0.65),arrowprops=dict(facecolor='g',alpha=0.3,shrink=0.1))
# xy represents the point, 

ax[1].annotate('End Points',xy=(0.2,0.1),xytext=(0.4,0.2),arrowprops=dict(alpha=0.3,shrink=0.1))
ax[1].annotate('End Points',xy=(0.5,0.4),xytext=(0.4,0.2),arrowprops=dict(alpha=0.3,shrink=0.1))
ax[1].annotate('End Points',xy=(0.8,0.1),xytext=(0.4,0.2),arrowprops=dict(alpha=0.3,shrink=0.1))
ax[1].plot([0.4,0.6],[0.9,0.8],'r.');
```

```python
# Twin Axes

fig,ax_c=plt.subplots(figsize=(11,6))
a=np.random.randint(-20,30,50)

def celsius_to_farenheit(temp_c):
    temp_f=((9*temp_c)//5)+32
    return temp_f

ax_c.set_xlabel("Days")
ax_c.tick_params(axis='x',bottom=False,labelbottom=False) # removing ticks from the x-axis

ax_c.plot(a)
ax_f=ax_c.twinx() # twin axis is created on the right side of the plot

ax_c.set_ylabel('$Temp\ (\degree C)$')
ax_f.set_ylabel('$Temp\ (F)$')

ymin,ymax=ax_c.get_ylim() # fetching min & max value for y-axis
ax_f.set_ylim(celsius_to_farenheit(ymin),celsius_to_farenheit(ymax)); # setting limits for the twin axis
```

### Box Plot

```python
fig=plt.figure(figsize=(18,12))

x=np.random.randint(0,20,50)
x=np.append(x,[35,40,25,30]) # introducing outliers

ax1=fig.add_subplot(231)
bp1=ax1.boxplot(x,showfliers=False) # showfliers removes the outliers from the plot

ax2=fig.add_subplot(232)
bp2=ax2.boxplot(x,vert=False,notch=True) # plots the graph horizontally, notch highlights the median

ax3=fig.add_subplot(233)
bp3=ax3.boxplot(x,patch_artist=True) # patch_artist introduces the patch in the plot

# styling individual components

ax4=fig.add_subplot(234)
bp4=ax4.boxplot(x,patch_artist=True)
bp4['boxes'][0].set(facecolor='green',alpha=0.4,edgecolor='blue',hatch='o')
bp4['medians'][0].set(linestyle='--',linewidth=3,color='maroon');

ax5=fig.add_subplot(235)
bp5=ax5.boxplot(x,patch_artist=True,vert=False)
bp5['boxes'][0].set(facecolor='pink',alpha=0.5,edgecolor='red',hatch='.')
bp5['whiskers'][0].set(linewidth=2,color='blue')
bp5['whiskers'][1].set(linewidth=2,color='green')

ax6=fig.add_subplot(236)
bp6=ax6.boxplot(x,patch_artist=True)
bp6['fliers'][0].set(marker='D',markerfacecolor='Red')
bp6['boxes'][0].set(facecolor='red',alpha=0.5,edgecolor='blue',hatch='||')
bp6['medians'][0].set(linestyle=':',linewidth=4,color='black');
```

### Autocorrelation Plot

```python
# Correlation: The measure of the relationship between two items or variables

# Autocorrelation: Measures the relationship between a variable's current value and past value.
# In Autocorrelation, same time series is used multiple times by lagging over one or more time periods
# The value ranges b/w [-1,1], 1: Perfect positive correlation, -1: Perfect negative correlation

# defining random data
plt.figure(figsize=(11,6))
data=np.random.randint(100,300,300).reshape(100,3)
d=pd.date_range(start="26th July 2010",end="13th Nov 2018",freq='M')
df=pd.DataFrame(data,index=d,columns=['Count_X','Count_Y','Count_Z'])

lags,c,vlines,hline=plt.acorr(df['Count_X'],normed=False,maxlags=20)
# normed=False doesn't normalizes the input vector, maxlags provides the max. number of lags
print("Lags:",lags)
print("\nc:",c)
print('\nvlines:',vlines)
print('\nhline:',hline)
```

### Expanding & Rolling Time (Plotting MA: Moving Averages)

```python
# initializing stock
plt.style.use("default")
tm=quandl.get("NSE/TATAMOTORS",start_date=pd.to_datetime('2017-09-1').date(), end_date=pd.to_datetime('2018-12-30').date(),\
              authtoken="YPB6c-2eBjnBvXtv7qB3")

# filtering & plotting stock 
dates=tm.loc['2017-09-1':'2018-10-14'].index # Filtered Dates
price=tm.loc['2017-09-1':'2018-10-14']['Close'] # Filtered Price for selected dates
plt.subplot()
plt.ylabel('Price')
plt.plot(dates,price,label='Price')

# expanding time (example: overall trend on average)
price.expanding().mean().plot(label='Overall Trend')
plt.legend();

# rolling time (example: moving mean averages)
tm['Close 30 day MA']=tm.rolling(window=30).mean()['Close'] # 30 days
tm['Close 90 day MA']=tm.rolling(window=90).mean()['Close'] # 90 days
tm[['Close','Close 30 day MA','Close 90 day MA']].plot(figsize=(16,6));
```

### Bollinger Bands (based on rolling time)

```python
tm['Close 20 day MA']=tm.rolling(window=20).mean()['Close'] # 20 MA
tm['Upper Band']=tm['Close 20 day MA'] + 2*(tm.rolling(window=20).std()['Close']) # 20 MA + 2*std(20)
tm['Lower Band']=tm['Close 20 day MA'] - 2*(tm.rolling(window=20).std()['Close']) # 20 MA - 2*std(20)

tm[['Close','Close 20 day MA','Upper Band','Lower Band']].plot(title='Bollinger Bands',figsize=(16,6));
```

```python
# Monthly Expenditure example

url='https://raw.githubusercontent.com/jamwine/Data/master/Monthly%20Expenditure.csv'
data=pd.read_csv(url)
df=pd.DataFrame(data)
all_categories=list(df['Category'].unique())
#df_grouped=df.groupby('Category')
a=df.groupby(['Category','Expenditure']).sum()
a.sort_values(['Cost'],ascending=False,inplace=True)

def make_plot(n=8,category='Food'):
    fig,ax=plt.subplots(figsize=(n,6))
    top=a.loc[category].head(n)
    top.plot(kind='bar',ax=ax,alpha=0.7)
    ax.set_ylabel('Cost')

interact(make_plot,n=(1,10),category=all_categories);
```
